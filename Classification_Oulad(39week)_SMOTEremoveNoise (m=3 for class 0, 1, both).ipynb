# Resampling - SMOTE - Remove Border


```python
from collections import Counter
from imblearn.over_sampling import SMOTE
```

```python
# hàm xác định các phần tử nhiễu và xóa
from sklearn.neighbors import NearestNeighbors
def remove_noise_class0(data, labels, m):

    # Tạo mô hình K-Nearest Neighbors
    knn = NearestNeighbors(n_neighbors=m + 1)
    knn.fit(data)

    # Tìm các láng giềng gần nhất
    neighbors = knn.kneighbors(data, return_distance=False)

    # Danh sách các cặp (dữ liệu, nhãn) giữ lại
    keep_data = []
    keep_labels = []

    for i, point_neighbors in enumerate(neighbors):
        # Bỏ qua chính điểm đó (index 0)
        neighbor_labels = labels[point_neighbors[1:]]

        if labels[i] == 0:  # Chỉ kiểm tra với lớp 0
            count_class_0 = np.sum(neighbor_labels == 0)
            count_class_1 = np.sum(neighbor_labels == 1)

            # Điều kiện giữ lại điểm
            if count_class_1 < m:  # Không phải là nhiễu
                keep_data.append(data[i])
                keep_labels.append(labels[i])
        else:
            # Giữ lại tất cả các điểm thuộc lớp 1
            keep_data.append(data[i])
            keep_labels.append(labels[i])

    # Chuyển về numpy arrays
    data_clean = np.array(keep_data)
    labels_clean = np.array(keep_labels)

    return data_clean, labels_clean
```

```python
#Xóa nhiễu lớp 0
X_clean0, y_clean0 = remove_noise_class0(X, y, m=3)

print("Dữ liệu sau khi xóa phần tử nhiễu và biên:")
print(X_clean0.shape)
print(y_clean0.shape)
```

```python
def remove_noise_class1(data, labels, m):

    # Tạo mô hình K-Nearest Neighbors
    knn = NearestNeighbors(n_neighbors=m + 1)
    knn.fit(data)

    # Tìm các láng giềng gần nhất
    neighbors = knn.kneighbors(data, return_distance=False)

    # Danh sách các cặp (dữ liệu, nhãn) giữ lại
    keep_data = []
    keep_labels = []

    for i, point_neighbors in enumerate(neighbors):
        # Bỏ qua chính điểm đó (index 0)
        neighbor_labels = labels[point_neighbors[1:]]

        if labels[i] == 1:  # Chỉ kiểm tra với lớp 0
            count_class_0 = np.sum(neighbor_labels == 0)
            count_class_1 = np.sum(neighbor_labels == 1)

            # Điều kiện giữ lại điểm
            if count_class_0 < m:  # Không phải là nhiễu
                keep_data.append(data[i])
                keep_labels.append(labels[i])
        else:
            # Giữ lại tất cả các điểm thuộc lớp 1
            keep_data.append(data[i])
            keep_labels.append(labels[i])

    # Chuyển về numpy arrays
    data_clean = np.array(keep_data)
    labels_clean = np.array(keep_labels)

    return data_clean, labels_clean
```

```python
#Xóa nhiễu lớp 1
X_clean1, y_clean1 = remove_noise_class1(X, y, m=3)

print("Dữ liệu sau khi xóa phần tử nhiễu và biên:")
print(X_clean1.shape)
print(y_clean1.shape)
```
